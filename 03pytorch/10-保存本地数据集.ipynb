```python
# 依赖引入
import torch
import os
import numpy as np
from torchvision import datasets, transforms
```

```python

# 创建保存目录
save_dir = './my_mnist_model'
os.makedirs(save_dir, exist_ok=True)

trainsform = transforms.Compose([transforms.ToTensor()])
train_data = datasets.MNIST(root='./data', train=True, download=True, transform=trainsform)
test_data = datasets.MNIST(root='./data', train=False, download=False, transform=trainsform)

# 提取训练数据
train_images = []
train_label = []

for i in range(len(train_data)):
    image, label = train_data[i]
    # 转转数据为numpy数组并去掉通道维度,squeeze:去掉图像通道
    image = (image.squeeze() * 255).numpy().astype(np.uint8)
    train_images.append(image)
    train_label.append(label)

# 将内容保存为numpy nparray,，虽然train_images/train_label中每个对象已经是numpy数组了，虽然通过numpy将图像从numpy tensor张量变为numpy array，但追加到空列中，还是原生空列表张量，所以还需要将原生列表转为nparray
train_images = np.array(train_images)
train_label = np.array(train_label)


# 提取测试数据
test_images = []
test_label = []

for i in range(len(test_data)):
    image, label = test_data[i]
    # 转转数据为numpy数组并去掉通道维度,squeeze:去掉图像通道
    image = (image.squeeze() * 255).numpy().astype(np.uint8)
    test_images.append(image)
    test_label.append(label)

# 将内容保存为numpy nparray,，虽然train_images/train_label中每个对象已经是numpy数组了，虽然通过numpy将图像从numpy tensor张量变为numpy array，但追加到空列中，还是原生空列表张量，所以还需要将原生列表转为nparray
test_images = np.array(test_images)
test_label = np.array(test_label)

# 保存成为numpy文件
np.save(os.path.join(save_dir, 'train_images.npy'), train_images)
np.save(os.path.join(save_dir, 'train_label.npy'), train_label)
np.save(os.path.join(save_dir, 'test_images.npy'), test_images)
np.save(os.path.join(save_dir, 'test_label.npy'), test_label)

```
