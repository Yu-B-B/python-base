```python
import torch
import numpy as np
```

```python
# 将numpy类型转为pytorch类型
np_array = np.array([[1,3], [4,9]])

py_array = torch.from_numpy(np_array)

print('np_array type is ', type(np_array))
print('tp_array type is ', type(py_array))

print('tp_array', py_array)
```

```python
# 将pytorch类型转为numpy类型
py_source_array = torch.tensor([[1,23,5],[5,2,12]])
np_change_array = py_source_array.numpy()

print('pytorch 类型转为 numpy 类型：')
print('原 py_source_array 类型为:', type(py_source_array))
print('切换后 np_change_array 类型为:', type(np_change_array))
print('切换后的数据为：', np_change_array)
```
