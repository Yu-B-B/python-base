```python
import torch
```

```python
a = torch.tensor(([3,4],[5,6]))
b = torch.tensor(([6,7], [8,9]))
print('tensor 加法：', a + b)
```

```python
# 减法
print('tensor 减法', a - b)
```

```python
# 乘法
print(a * b)
```

```python
# 除法
print(a / b)
```

```python
# 张量矩阵乘法1（行 * 列）
print( torch.matmul(a, b))
```

```python
# 张量矩阵乘法2（全量）
print(a @ b)
```

```python
# 张量拼接 cat（dim = 0 按照第一个维度，行拼接，dim = 1 按照第二个维度，列拼接）
# dim 为dimension，dim若为-1时，表示为最后一个，若为-2，则为倒数第二个
print(torch.cat((a,b), dim = 1))

print(torch.cat((a,b), dim = -2))
```
